```python
training_data = []


def create_training_data():
    
   
    for category in CATEGORIES:
       
        path = os.path.join(DATADIR,category)  # create path to dogs and cats
        class_num = CATEGORIES.index(category)  # get the classification  (0 or a 1). 0=dog 1=cat
        
        for img in tqdm(os.listdir(path)):
            d = []
            l = img
            d.append(l)
           
             
            
            
            try:
                
                
                img_array = cv2.imread(os.path.join(path,img) ,cv2.IMREAD_UNCHANGED)  # convert to array
                im = cv2.bilateralFilter(np.array( img_array,dtype=np.uint8),9,70,70)
                imb = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
                im = cv2.adaptiveThreshold(imb, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 35 ,3)
                
                training_data.append([im, class_num])  # add this to our training_data
            except Exception as e:  # in the interest in keeping the output clean...
                pass
          
            
            file_name_path = str(d)
#            print(file_name_path)
#            break
            
            cv2.imwrite("new_data/"+file_name_path,im)
create_training_data()

print(len(training_data))
```
